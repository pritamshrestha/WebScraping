```python
# Web scraping with BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
# sending the requests to the url
url="https://www.worldometers.info/coronavirus/?utm_campaign=homeAdvegas1?"
r=requests.get(url)
# Printing all the text associated with that page!
print(r.text)
```

```python
# Using beautifulsoup parseing data
soup=BeautifulSoup(r.text,'html.parser')
results=soup.find_all('div',attrs={'class':'col-md-8'})
print(results)
```
